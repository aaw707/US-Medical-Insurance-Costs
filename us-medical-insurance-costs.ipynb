# U.S. Medical Insurance Costs

```python
import csv
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
print(df.head())
```

```python
average_age = np.mean(df['age'])
print('The average age of the patients is ' + str(round(average_age)) + ' years old.')
```

```python
plt.hist(df['age'], bins = 15, range = (15, 65))
ax = plt.subplot()
ax.set_xticks(range(15, 66, 5))
ax.set_xticklabels(['15', '20', '25', '30', '35', '40', '45', '50', '55', '60', '65'])
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title('Age Distribution among Patients')
plt.show()
```

```python
having_children = df[df.children > 0]
average_age_having_children = np.mean(having_children['age'])
print('The average age of people having at least one child is ' + str(round(average_age_having_children)) + ' years old.')
```

```python
children_count = df.groupby('children').region.count().reset_index()
children_count.rename(columns = {'region': 'count'}, inplace = True)
children_x = range(6)
children_y = children_count['count']
print(children_count)
```

```python
plt.bar(children_x, children_y)
plt.xlabel('Number of Children')
plt.ylabel('Number of Patients')
plt.title('Number of Children Owned by Patients')
plt.show()
```

```python
charge_by_region = df.groupby('region').charges.mean().reset_index()
plt.bar(range(4), charge_by_region.charges)
ax = plt.subplot()
ax.set_xticks(range(4))
ax.set_xticklabels(['Northeast', 'Northwest', 'Southeast', 'Southwest'])
plt.xlabel('Regions')
plt.ylabel('Average Charges')
plt.title('Average Insurance Charges at Each Region')
plt.show()
```

```python
patients_by_region = df.groupby('region').charges.count().reset_index()
patients_by_region.rename(columns = {'charges': 'count'}, inplace = True)
print(patients_by_region)
plt.bar(range(4), patients_by_region['count'])
ax = plt.subplot()
ax.set_xticks(range(4))
ax.set_xticklabels(['Northeast', 'Northwest', 'Southeast', 'Southwest'])
plt.xlabel('Regions')
plt.ylabel('Patient Counts')
plt.title('Patient Counts at Each Region')
plt.show()
```

```python
smokers = df[df['smoker'] == 'yes']
non_smokers = df[df['smoker'] == 'no']
smokers_min = smokers.charges.min()
smokers_max = smokers.charges.max()
smokers_mean = smokers.charges.mean()
non_smokers_min = non_smokers.charges.min()
non_smokers_max = non_smokers.charges.max()
non_smokers_mean = non_smokers.charges.mean()
print('The minimum insurance charge for smokers is ' + str(smokers_min) + ' dollars.')
print('The maximum insurance charge for smokers is ' + str(smokers_max) + ' dollars.')
print('The average insurance charge for smokers is ' + str(smokers_mean) + ' dollars.')
print('The minimum insurance charge for non smokers is ' + str(non_smokers_min) + ' dollars.')
print('The maximum insurance charge for non smokers is ' + str(non_smokers_max) + ' dollars.')
print('The average insurance charge for non smokers is ' + str(non_smokers_mean) + ' dollars.')
```

```python
males = df[df['sex'] == 'male']
females = df[df['sex'] == 'female']
print(males.bmi.max())
print(males.bmi.min())
print(females.bmi.max())
print(females.bmi.min())
plt.hist(males['bmi'], histtype = 'step', label = 'males')
plt.hist(females['bmi'], histtype = 'step', label = 'females')
plt.legend()
plt.xlabel('BMI')
plt.ylabel('Number of Patients')
plt.title('BMI across Genders')
plt.show()
```
